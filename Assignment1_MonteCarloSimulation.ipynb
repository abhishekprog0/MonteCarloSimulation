```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import plotly.express as px

sns.set()
size = 18
params = {'legend.fontsize': 'large',
          'figure.figsize': (20,7),
          'axes.labelsize': size * 0.8,
          'axes.titlesize': size * 0.95,
          'xtick.labelsize': size*0.75,
          'ytick.labelsize': size*0.75,
          'axes.titlepad': 25}
plt.rcParams.update(params)
```

```python
#parameter list for simulation
alpha_list = [0.4, 1.5, 3.3, 10]
beta_list = [0.6, 1.8, 3.5, 9]
N = 100000
```

### Algorithm 1

```python
X = np.zeros((N))
pdf = np.zeros((N))
for i in range(len(alpha_list)):
    
    for j in range(N):

        u1 = np.random.uniform(0.0, 1.0)
        u2 = np.random.uniform(0.0, 1.0)

        if u1 < 0.5:
            X[j] = (np.power(u2, 1.0/alpha_list[i]))

        else:
            X[j] = (1.0 - np.power(1.0 - u2, 1.0/beta_list[i]))

    print ("\n")
    values, bins, _ = plt.hist(X, bins=80, density = True, color='orange', alpha=0.55)
    plt.title("N = " + str(N) + ", Alpha = " + str(alpha_list[i]) + ", Beta = " + str(beta_list[i]))
    plt.xlabel("X")
    plt.ylabel("PDF of X")
    plt.show()
    print ("\n")
    area = sum(np.diff(bins)*values)
    #print ("Area under the histogram: " + str(round(area, 1)))
```
